# Outlier treatment

```python
# Importing the zscore and then finding out the zscores values in 'rotational speed' column, setting the threshold at 3 
# and then removing the outliers beyond the thershold value 
from scipy.stats import zscore
z_scores = zscore(df['Rotational speed'])
z_score_outliers=(z_scores<-3)|(z_scores>3)
z_score_outlier_rows=df[z_score_outliers]
print("outliers detected by Z-score:",z_score_outlier_rows)
```

```python
x=(z_scores>-3)&(z_scores<3)
new_df=df[x]
print(new_df)
```

```python
z_scores = zscore(new_df['Torque'])
z_score_outliers=(z_scores<-3)|(z_scores>3)
z_score_outlier_rows=new_df[z_score_outliers]
print("outliers detected by Z-score:",z_score_outlier_rows)
```

```python
x=(z_scores>-3)&(z_scores<3)
df_new=new_df[x]
print(df_new)
```

# Linear Regression

```python
# importing all the necessary libraries used in linear regression model
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
# Manually converting the categorical column into numerical column using the replace method
df_new.replace({'Type':{'L':0,'M':1,'H':2}},inplace=True)
```

```python
# Dropping two columns which are not useful in building the regression or classification model.
df_new.drop(['UDI','Product ID'],axis=1,inplace=True)
```

```python
# Using Standard scaler technique to convert all the numerical values into same decimal format
numerical_cols=['Type','Air temperature','Process temperature','Rotational speed','Torque',
               'Machine failure','TWF','HDF','PWF','OSF','RNF']
scaler=StandardScaler()
scaled_cols=pd.DataFrame(scaler.fit_transform(df_new[numerical_cols]),columns=scaler.get_feature_names_out(numerical_cols))
```

```python
scaled_cols
```

```python
# giving X and y values 
X=scaled_cols
Y=df_new['Tool wear']
```

```python
# splitting the X and Y into training and testing sets.
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# creating linear regression model in order to train the X and Y on the same and predicting the y values.
model_lr=LinearRegression()
model_lr.fit(X_train,Y_train)
y_pred=model_lr.predict(X_test)
```

```python
print(model_lr.intercept_) #y-intercept of the model
```

```python
print(model_lr.coef_) #slope values for each independent variable
```

```python
# Calculating performance metrics for linear regression model
mae = mean_absolute_error(Y_test,y_pred)
mse= mean_squared_error(Y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, y_pred)

print('Mean Absolute Error',mae)
print('Mean Squared Error',mse)
print('Root Mean Absolute Error',rmse)
print('R2 Score',r2)
```

```python
#adjusted_r2=1-[(1-r2)*(n-1)/(n-k-1)]
adjusted_r2=1-((1-0.042777)*(9345-1)/(9345-11-1))
print('adjusted r2 is :',adjusted_r2)
```

```python
# Calculating sum of squared residual,sum of squared total,sum of squared error for the model
y_mean=np.mean(Y_test)
SSR = np.sum((y_pred - y_mean) ** 2)
print('SSR :',SSR)
SST = np.sum((Y_test - y_mean) ** 2)
print('SST :',SST)
SSE=SST-SSR
print('SSE :',SSE)
```

# Logistic Regression


```python
# Using Standard scaler technique to convert all the numerical values into same decimal format
numerical_cols=['Type','Air temperature','Process temperature','Rotational speed','Torque',
               'Tool wear','TWF','HDF','PWF','OSF','RNF']
scaler=StandardScaler()
scaled_cols=pd.DataFrame(scaler.fit_transform(df_new[numerical_cols]),columns=scaler.get_feature_names_out(numerical_cols))
```

```python
# giving X and y values 
X=scaled_cols
Y=df_new['Machine failure']
```

```python
#importing the library for oversampling 
from imblearn.combine import SMOTETomek
```

```python
#performing oversampling using SMOTE
smk = SMOTETomek(random_state=42)
X_res,Y_res=smk.fit_resample(X,Y)
```

```python
#Oversampled data
X_res.shape,Y_res.shape
```

```python
# Importing all the important libraries 
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.utils.fixes import loguniform
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Splitting the data into training and testing data, training the dataset with logistic regression
X_train,X_test,Y_train,Y_test=train_test_split(X_res,Y_res,test_size=0.2,random_state=75)
log=LogisticRegression()
log.fit(X_train,Y_train)
```

```python
# Train and test scores are calculated
print("Train score",log.score(X_train,Y_train))
print("Test score",log.score(X_test,Y_test))
```

```python
#prediting the train and test values
pred_train = log.predict(X_train)
pred_test = log.predict(X_test)
```

```python
acc_log=accuracy_score(Y_test,pred_test)
acc_log
```

```python
#the classification report is generated on the training values
print(metrics.classification_report(Y_train,pred_train))
```

```python
#the classification report is generated on the testing values
print(metrics.classification_report(Y_test,pred_test))
```

```python
#AUC-ROC curve is plotted 
roc=log.predict_proba(X_train)[:,1]
fpr, tpr, threshold = metrics.roc_curve(Y_train,roc)
plt.plot([0,1], [0,1],'k--') #x=(0,1) y
plt.plot(fpr,tpr,label='logistic')
plt.ylabel("TPR")
plt.xlabel("FPR")
plt.show()
```

```python
metrics.roc_auc_score(Y_train,roc)
```

```python
from sklearn.metrics import matthews_corrcoef

mcc= matthews_corrcoef(Y_test,pred_test)
print("MCC: ",mcc)
```

```python
cm = confusion_matrix(Y_test,pred_test)
print('Confusion Matrix: ',cm)
sns.heatmap(cm, annot=True, fmt = '.3g')
plt.title('Confusion Matrix - Test Data')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
